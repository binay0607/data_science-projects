```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('home_prices_with_townname.csv')
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final=merged.drop(['town','jharsuguda'],axis='columns') #as we can get the other one based on the available two also it makes dummies trap so dropping one dummy is a thumb rule
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x=final.drop('price',axis='columns')
x
```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])   # 2800sqft in sambalpur
```

```python
model.predict([[2800,0,0]]) 
```

```python
model.score(x,y)
```
